```python
def load_data(data, name):
    df = sio.loadmat(data) ## Reference the data location
    df = pd.DataFrame(np.hstack(df[name][:,0]))

    return df
        
```

```python
def extract(data, name, *args):
    df = load_data(data, name)
    y = []
    d = {}
    for arg in args:
        d[arg] = []
        for i in range(0, len(df[arg][0])):
            d[arg].append(df[arg][0][i][0])
        
    return pd.DataFrame(d)
```

```python
## Should fix weird timestamps
from datetime import datetime, timedelta
def hour_rounder(t):
    # Rounds to nearest hour by adding a timedelta hour if minute >= 30
    return (t.replace(second=0, microsecond=0, minute=0, hour=t.hour)
               +timedelta(hours=t.minute//30))
```

```python
def threshold(df):
    ordered_rain = df['precip'].sort_values(ascending=True)
    new_sum = np.array(ordered_rain).cumsum()
    scaled_sum = new_sum/new_sum.max()
    first = len(scaled_sum[scaled_sum >= 0.5])
    second = len(scaled_sum[scaled_sum >= 0.25])

    thres = [2 if i >= ordered_rain[-first:(-first +1)].values else 1 if i >= ordered_rain[-second:(-second +1)].values else 0 for i in df['precip']]
    return thres
```

```python
def plot_2d_space(X, y, label='Classes'):   
    colors = ['#1F77B4', '#FF7F0E', 'purple']
    markers = ['o', 's', '>']
    plt.figure(figsize=(8,8))
    for l, c, m in zip(np.unique(y), colors, markers):
        plt.scatter(
            X[y==l, 0],
            X[y==l, 1],
            X[y==l, 2],
            c=c, label=l, marker=m
        )
    plt.title(label)
    plt.legend(loc='upper right')
    plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
X = pca.fit_transform(X)

plot_2d_space(X, y, 'Imbalanced dataset (2 PCA components)')
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(return_indices=True)
X_rus, y_rus, id_rus = rus.fit_sample(X, y)

print('Removed indexes:', id_rus)

plot_2d_space(X_rus, y_rus, 'Random under-sampling')
```

```python
from imblearn.under_sampling import TomekLinks

tl = TomekLinks(return_indices=True, ratio='majority')
X_tl, y_tl, id_tl = tl.fit_sample(X, y)

print('Removed indexes:', id_tl)

plot_2d_space(X_tl, y_tl, 'Tomek links under-sampling')
```

```python
from imblearn.under_sampling import ClusterCentroids

cc = ClusterCentroids(ratio={0: 10})
X_cc, y_cc = cc.fit_sample(X, y)

plot_2d_space(X_cc, y_cc, 'Cluster Centroids under-sampling')
```

```python
from imblearn.combine import SMOTETomek

smt = SMOTETomek(ratio='auto')
X_smt, y_smt = smt.fit_sample(X, y)

plot_2d_space(X_smt, y_smt, 'SMOTE + Tomek links')
```

```python
def attempt(data):
    data['freq'] = [0.6409 if i==2 else 0.359 if i==1 else 0.0001 for i in data['thres']]
    data['no_rain'] = [1 if i == 0 else 0 for i in data['precip']]
    data['count'] = data.groupby((data['no_rain'] != data['no_rain'].shift(1)).cumsum()).cumcount()
    df = data.sample(frac=0.3, weights = data.freq)
    
    class_weights = list(class_weight.compute_class_weight('balanced',
                                             np.unique(df['thres']),
                                             df['thres']))

    
        
    X = df[['sss','sst','wind_spd','sss_change','wind_change','differential','temp_change','count']]
    y=df['thres']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
    
    w_array = np.ones(y_train.shape[0], dtype = 'float')
    for i, val in enumerate(y_train):
        w_array[i] = class_weights[val-1]
    
    xgb = XGBClassifier(objective='multi:softmax', num_class=3, random_state=42, tree_method='hist', learning_rate=0.1,
                                   n_estimators=500, booster='gbtree', grow_policy='lossguide',
                                   colsample_bytree=0.8)

    xgb.fit(X_train, y_train, sample_weight=w_array)
    pred = xgb.predict(X_test)
    print(confusion_matrix(y_test, pred))
    print(classification_report(y_test, pred))
```
